### 재료명 컬럼 정제

```python
all_igd = new_recipe['CKG_MTRL_CN']
all_igd
```

```python
# 정제1) split 함수
def Split(txt,dmter):
    slt = txt.split(dmter)
    slt = [v for v in slt if v]
    return slt
```

```python
# 재료명 컬럼 정제 후 -> temp_list: 재료내용,재료명,중량,중량숫자만,중량단위,재료분류
temp_list = []
append = temp_list.append
for ingre in tqdm(all_igd, desc='progress: ', mininterval=0.01):
    slt1 = Split(ingre, '[')
    for slt2 in slt1:
        p = re.compile(r"(.+)\](.+)")
        m1 = p.match(slt2)
        if m1:
            cate_name = m1.group(1)
    #         print(cate_name)
            igd_group = m1.group(2)
    #         print(igd_group)
            slt2 = Split(igd_group,'|')
            for igds in slt2:
                p = re.compile('\d\/\d|\d\-\d|\d\~\d|\d\+\d\/\d|\d+.\d+|\d+|약간|적당량')
                m2 = p.search(igds)
                if m2:
                    idx = m2.start()
                    igd = igds[:idx].strip()
                    weight = igds[idx:].strip()
                    weight_num = m2.group()
                    unit = re.compile('\d\/\d|\d\-\d|\d\~\d|\d\+\d\/\d|\d+.\d+|\d+|약간|적당량').sub('',weight)
                else:
                    igd = igds.strip()
                    weight = '-'
                    weight_num = '-'
                    unit = '-'
                append([ingre,igd,weight,weight_num,unit,cate_name])
        else:
            continue
```

```python
print('재료명 수:', len(temp_list), '개')
```

```python
# temp_list df로 만들기
col_name = ['CKG_MTRL_CN', 'ingredients', 'weight', 'weight_num', 'unit', 'cate_name']
list1 = temp_list
list_df = pd.DataFrame(list1, columns=col_name)
list_df.head()
```

```python
# 요리명_컬럼 + 재료명_df 합치기
merge_df = pd.merge(new_recipe, list_df)
merge_df
```
